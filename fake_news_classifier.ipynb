# Fake News Classifier Using LSTM

```python
# Import Dependencies
import tensorflow as tf 
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Embedding
```

```python
import pandas as pd 

df = pd.read_csv("fake_news.csv")
df.head()
```

```python
# Check the null values
df.isnull().sum()
```

```python
# Drop Nan values
df = df.dropna()
```

```python
# Get th Independent Features and dependent feature
X = df.drop("label" , axis=1)
y = df.label
```

```python
X.shape , y.shape
```

## One-Hot Representation

```python
# Vocabulary size
voc_size=5000
```

```python
messages = X.copy()
```

```python
messages["title"][1]
```

```python
messages["text"][0]
```

```python
messages.reset_index(inplace=True)
```

```python
# import Dependencies for NLP task
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download("stopwords")
```

```python
# Data Preprocessing
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    print(i)
    review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
onehot_repr=[one_hot(words,voc_size)for words in corpus] 
onehot_repr
```

## Embedding Representation

```python
sent_length = 20
embedded_docs = pad_sequences(onehot_repr , padding="pre" , maxlen = sent_length)
print(embedded_docs)
```

```python
embedded_docs[0]
```

```python
# Creating model
embedding_vector_features = 40
model = Sequential()
model.add(Embedding(voc_size , embedding_vector_features , input_length = sent_length))
model.add(LSTM(100)) # one LSTM layer with 100 neurons
model.add(Dense(1 , activation = "sigmoid"))
```

```python
model.summary()
```

```python
model.compile(loss="binary_crossentropy" , optimizer = "adam" , metrics=["accuracy"])
```

```python
len(embedded_docs) , y.shape
```

```python
import numpy as np 
X_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
X_final.shape , y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X_final , y_final , test_size=0.33 , random_state=42)
```

## Training the model

```python
model.fit(X_train , y_train , validation_data=(X_test , y_test) , epochs=10 , batch_size = 64)
```

## Check the Metrics and Accuracy

```python
y_pred = model.predict_classes(X_test)
```

```python
y_test[:10]
```

```python
y_pred[:10]
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

### If we are adding the Dropout layer 

```python
from tensorflow.keras.layers import Dropout
## Creating model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
y_pred=model.predict_classes(X_test)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

Accuracy is much more better if we add the Dropout layer
